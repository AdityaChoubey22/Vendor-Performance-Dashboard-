```python
import os
import pandas as pd
from sqlalchemy import create_engine
import logging
import time 

logging.basicConfig(filename="logs/ingestion_db.log",
                   level=logging.DEBUG,
                   format="%(asctime)s-%(levelname)s-%(message)s ",
                   filemode="a"
                   )

engine = create_engine("sqlite:///inventory.db")


def ingest_db(df,table_name,engine):
    #this fxn will ingest the dataframe into database table
    df.to_sql(table_name,con=engine,if_exists="replace",index= False)


#def load_raw_data():
    # this function will load the csv as data frame and ingest it in database db
   # start = time.time()
    #for file in os.listdir("data"):
     #   if file.endswith(".csv") and "ipynb_checkpoints" not in file:
      #      df = pd.read_csv("data/" + file)
       #     if len(df) > 2000000:
        #        df = df.sample(n=2000000, random_state=42)

         #   logging.info(f"ingesting {file} in db with {len(df)} rows")
          #  ingest_db(df, file[:-4], engine)
   # end=time.time() 
    #total_time =(end - start)/60
  #  logging.info('____________Ingestion Complete_____________')    
   # logging.info(f"\nTotal Time Taken:{total_time} minutes")         
    
                       ###only to read .csv file     
                        ### but not to read .ipynb_checkpoints

def load_raw_data():
    data_dir = "data"
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)
        logging.warning(f"Data directory '{data_dir}' not found. Created an empty one.")
        return  # no files to process yet

    start = time.time()
    for file in os.listdir(data_dir):
        if file.endswith(".csv") and "ipynb_checkpoints" not in file:
            file_path = os.path.join(data_dir, file)
            try:
                logging.info(f"Started ingestion for {file}")
                df = pd.read_csv(file_path)
                if len(df) > 2000000:
                    df = df.sample(n=2000000, random_state=42)

                ingest_db(df, file[:-4], engine)
                logging.info(f"Completed ingestion for {file} with {len(df)} rows")
            except Exception as e:
                logging.error(f"Error processing {file}: {e}")

    end = time.time()
    total_time = (end - start) / 60
    logging.info("____________Ingestion Complete_____________")
    logging.info(f"Total Time Taken: {total_time:.2f} minutes")    
if __name__=='__main__':                  
    load_raw_data()
```
